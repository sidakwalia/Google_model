# Text ads and expanded text ads from google_ads_insights

```python
def text_ads_exp(date_range,metric,account_id):
    daily_insight_query=db.table('google_ads_insights').select('ad_id','spend','insights','conversions')
    daily_ads=daily_insight_query.where('account_id',account_id).where_between('date', date_range).get()
    ads_daily=pd.DataFrame(list(daily_ads))
    ads_daily = ads_daily.fillna(0)
    ads_daily['spend'] = ads_daily['spend'].astype(float)
    ads_daily['conversions'] = ads_daily['conversions'].astype(float)
    df=ads_daily.copy()
    google_ads=db.table('google_ads').select('id','ad_type')
    google_ads=google_ads.where('account_id',account_id).where_in('ad_type',['EXPANDED_TEXT_AD','TEXT_AD']).get()
    google_ads=pd.DataFrame(list(google_ads))
    google_ads_list=list(google_ads['id'])  
    df=df[df['ad_id'].isin(google_ads_list)]
    ads_list = df['ad_id'].unique()
    return df,ads_list
```

```python
def cal(df,df1,ads_list):
    for ad_id in ads_list:
        super_dic={}
        print("ad_id---------------------------------------------",ad_id)
        ad_spend=0
        ad_conversions=0
        ad_insight=df[df['ad_id']==ad_id]['insights']
        ad_spend += df[df['ad_id'] == ad_id]['spend'].sum()
        ad_conversions+=df[df['ad_id'] == ad_id]['conversions'].sum()
        for insight in ad_insight:
            for k, v in json.loads(insight).items():
                super_dic.setdefault(k, []).append(v)
        for k, v in super_dic.items():
            if k in sum_insight:
                if type(v) == list:
                    # print(v,"key is---------",k)
    #                 if any('.' in n for n in v):
                    super_dic[k] = list(map(float, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])
                else:
                    super_dic[k] = list(map(int, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])
            elif k in avg_insight:
                if type(v) == list:
                    # print(v,"key is---------",k)
                    # if any('.' in n for n in v):
                    super_dic[k] = list(map(float, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])/len(super_dic[k])
                else:
                    super_dic[k] = list(map(int, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])/len(super_dic[k])
        if super_dic['impressions']!=0:
            super_dic['ctr']=(super_dic['clicks']/super_dic['impressions'])*100
        else:
            super_dic['ctr']=0
        super_dic['spend']=ad_spend
        super_dic['conversions']=ad_conversions
        if super_dic['conversions']!=0:
            super_dic['cost_per_conversions']=(super_dic['spend']/super_dic['conversions'])
        else:
            super_dic['cost_per_conversions']=0
        try:
            print("the number is------------------------------",df[df['ad_id']==ad_id].index[0])
            df1['insights'][df1[df1['ad_id']==ad_id].index[0]] = json.dumps((super_dic))
        except:
            print("Not found ad_id ", ad_id, 'in ads_preprocessed')
    return df1
```

```python
def result_cal(df,flag,metric):
    df['clicks'] = df['insights'].apply(lambda x: float(json.loads(x)['clicks']) if 'clicks' in json.loads(x).keys() else 0)
    df['spend']=df['insights'].apply(lambda x: float(json.loads(x)['spend']) if 'spend' in json.loads(x).keys() else 0)
    df['impressions'] = df['insights'].apply(lambda x: float(json.loads(x)['impressions']) if 'impressions' in json.loads(x).keys() else 0)
    df['conversions'] = df['insights'].apply(lambda x: float(json.loads(x)['conversions']) if 'conversions' in json.loads(x).keys() else 0)
    df['ctr']=(df['clicks']/df['impressions'])*100
    df['cost_per_conversions'] = df['insights'].apply(lambda x: float(json.loads(x)['cost_per_conversions']) if 'cost_per_conversions' in json.loads(x).keys() else 0)
    df[metric] = df['insights'].apply(lambda x: json.loads(x)[metric] if metric in json.loads(x).keys() else None)
    df[metric] = pd.to_numeric(df[metric], errors='coerce')
    df = df[pd.notna(df[metric])]
    if flag=='dynamic':
        cols=['account_id','value','ctr','impressions','clicks','spend','cost_per_conversions','conversions','type']
    elif flag=='normal':
        cols=['account_id','title','ctr','impressions','clicks','body','spend','cost_per_conversions','conversions','ad_type']
    df['main_json'] = df[cols].apply(lambda x: x.to_json(), axis=1)
    top_worst_text_json=top_and_worst_text(df,metric,flag)
    return df,top_worst_text_json

def top_and_worst_text(df,metric,flag):
    print("Inside top_worst")
    size1 = len(df)
    top_size = int(size1/2)
    worst_size = int(size1/2)
    top_text_lst = []
    worst_text_lst = []
    df.sort_values(by=metric, ascending=False)

    top_perfomning_ad = list(df.sort_values(
        by=metric, ascending=False).head(top_size)['ad_id'])
    worst_perfomning_ad = list(df.sort_values(
        by=metric, ascending=False).tail(worst_size)['ad_id'])
    for i in worst_perfomning_ad:
        worst_text_lst.append(i)
    for i in top_perfomning_ad:
        top_text_lst.append(i)
    text={}
    top_text,worst_text={},{}
    if flag=='normal':
        df_top_text=df[df['ad_id'].isin(top_text_lst)]
        for ind in df_top_text.index:
            ad_id_tit=df_top_text['ad_id'][ind]+"_title"
            ad_id_body=df_top_text['ad_id'][ind]+"_body"
            top_text[ad_id_tit]=json.loads(df_top_text['main_json'][ind])    
            top_text[ad_id_body]=json.loads(df_top_text['main_json'][ind])
        df_worst_text=df[df['ad_id'].isin(worst_text_lst)]
        for ind in df_worst_text.index:
            ad_id_tit=df_worst_text['ad_id'][ind]+"_title"
            ad_id_body=df_worst_text['ad_id'][ind]+"_body"
            worst_text[ad_id_tit]=json.loads(df_worst_text['main_json'][ind])    
            worst_text[ad_id_body]=json.loads(df_worst_text['main_json'][ind])
        text['top_text']=top_text
        text['worst_text']=worst_text        
        return text
    elif flag=='dynamic':
        df_top_text=df[df['ad_id'].isin(top_text_lst)]
        for ind in df_top_text.index:
            ad_id_tit=df_top_text['ad_id'][ind]+"_asset"
            top_text[ad_id_tit]=json.loads(df_top_text['main_json'][ind])    
        df_worst_text=df[df['ad_id'].isin(worst_text_lst)]
        for ind in df_worst_text.index:
            ad_id_tit=df_worst_text['ad_id'][ind]+"_asset"
            worst_text[ad_id_tit]=json.loads(df_worst_text['main_json'][ind])    
        text['top_text_asset']=top_text
        text['worst_text_asset']=worst_text        
        return text

```

```python
def asset_calculation(date_range,metric,account_id):
    try:
        asset_query_insights=db.table('google_ads_asset_mapping').select('ad_id','asset_id','spend','insights','conversions','field_type','account_id')
        asset_query_insights=asset_query_insights.where('account_id',account_id).where_in('field_type',["HEADLINE","DESCRIPTION"]).where_between('date', date_range).get()
        asset_query_insights=pd.DataFrame(list(asset_query_insights))
        asset_query_insights = asset_query_insights.fillna(0)
        asset_query_insights['spend'] = asset_query_insights['spend'].astype(float)
        asset_query_insights['conversions'] = asset_query_insights['conversions'].astype(float)
        asset_query_insights_dup=asset_query_insights[['ad_id','spend','insights','conversions','asset_id','field_type','account_id']]
        asset_ads_list = asset_query_insights_dup['ad_id'].unique()
        return asset_query_insights_dup,asset_ads_list
    except:
        print("Less Data------------------------------>")
        response={"data":{},"status":False}
        return response
```

```python
def get_clean_data(x):
    temp = [re.sub('[0-9]', '', i).replace("_","").replace("''","").replace('""',"") for i in x] 
    temp = [i for i in temp if len(i) > 2]
    return temp
def get_type_feature(x,key):
    if str(x).lower() == key:
        return 1
    else:
        return 0
def key_mapping_to_value(x, key):
    try:
        if key in x:
            return 1
        else:
            return 0
    except Exception as e:
        return 0
    
def set_valid_keys(x):
    try:
        return list(set([i.lower().replace("segment_label_","").replace("shot_label_","") for i in list(ast.literal_eval(str(x)).keys())]))
    except Exception as e:
        return []
```

```python
def result_cal(df,flag,metric):
    df['clicks'] = df['insights'].apply(lambda x: float(json.loads(x)['clicks']) if 'clicks' in json.loads(x).keys() else 0)
    df['spend']=df['insights'].apply(lambda x: float(json.loads(x)['spend']) if 'spend' in json.loads(x).keys() else 0)
    df['impressions'] = df['insights'].apply(lambda x: float(json.loads(x)['impressions']) if 'impressions' in json.loads(x).keys() else 0)
    df['conversions'] = df['insights'].apply(lambda x: float(json.loads(x)['conversions']) if 'conversions' in json.loads(x).keys() else 0)
    df['ctr']=(df['clicks']/df['impressions'])*100
    df['cost_per_conversions'] = df['insights'].apply(lambda x: float(json.loads(x)['cost_per_conversions']) if 'cost_per_conversions' in json.loads(x).keys() else 0)
    df[metric] = df['insights'].apply(lambda x: json.loads(x)[metric] if metric in json.loads(x).keys() else None)
    df[metric] = pd.to_numeric(df[metric], errors='coerce')
    df = df[pd.notna(df[metric])]
    if flag=='dynamic':
        cols=['account_id','value','ctr','impressions','clicks','spend','cost_per_conversions','conversions','type']
    elif flag=='normal':
        cols=['account_id','title','ctr','impressions','clicks','body','spend','cost_per_conversions','conversions','ad_type']
    df['main_json'] = df[cols].apply(lambda x: x.to_json(), axis=1)
    top_worst_text_json=top_and_worst_text(df,metric,flag)
    return top_worst_text_json

def top_and_worst_text(df,metric,flag):
    print("Inside top_worst")
    size1 = len(df)
    top_size = int(size1/2)
    worst_size = int(size1/2)
    top_text_lst = []
    worst_text_lst = []
    df.sort_values(by=metric, ascending=False)

    top_perfomning_ad = list(df.sort_values(
        by=metric, ascending=False).head(top_size)['ad_id'])
    worst_perfomning_ad = list(df.sort_values(
        by=metric, ascending=False).tail(worst_size)['ad_id'])
    for i in worst_perfomning_ad:
        worst_text_lst.append(i)
    for i in top_perfomning_ad:
        top_text_lst.append(i)
    text={}
    top_text,worst_text={},{}
    if flag=='normal':
        df_top_text=df[df['ad_id'].isin(top_text_lst)]
        for ind in df_top_text.index:
            ad_id_tit=df_top_text['ad_id'][ind]+"_title"
            ad_id_body=df_top_text['ad_id'][ind]+"_body"
            top_text[ad_id_tit]=json.loads(df_top_text['main_json'][ind])    
            top_text[ad_id_body]=json.loads(df_top_text['main_json'][ind])
        df_worst_text=df[df['ad_id'].isin(worst_text_lst)]
        for ind in df_worst_text.index:
            ad_id_tit=df_worst_text['ad_id'][ind]+"_title"
            ad_id_body=df_worst_text['ad_id'][ind]+"_body"
            worst_text[ad_id_tit]=json.loads(df_worst_text['main_json'][ind])    
            worst_text[ad_id_body]=json.loads(df_worst_text['main_json'][ind])
        text['top_text']=top_text
        text['worst_text']=worst_text        
        return text
    elif flag=='dynamic':
        df_top_text=df[df['ad_id'].isin(top_text_lst)]
        for ind in df_top_text.index:
            ad_id_tit=df_top_text['ad_id'][ind]+"_asset"
            top_text[ad_id_tit]=json.loads(df_top_text['main_json'][ind])    
        df_worst_text=df[df['ad_id'].isin(worst_text_lst)]
        for ind in df_worst_text.index:
            ad_id_tit=df_worst_text['ad_id'][ind]+"_asset"
            worst_text[ad_id_tit]=json.loads(df_worst_text['main_json'][ind])    
        text['top_text_asset']=top_text
        text['worst_text_asset']=worst_text        
        return text

```

```python
def cal(df,df1,ads_list):
    for ad_id in ads_list:
        super_dic={}
        print("ad_id---------------------------------------------",ad_id)
        ad_spend=0
        ad_conversions=0
        ad_insight=df[df['ad_id']==ad_id]['insights']
        ad_spend += df[df['ad_id'] == ad_id]['spend'].sum()
        ad_conversions+=df[df['ad_id'] == ad_id]['conversions'].sum()
        for insight in ad_insight:
            for k, v in json.loads(insight).items():
                super_dic.setdefault(k, []).append(v)
        for k, v in super_dic.items():
            if k in sum_insight:
                if type(v) == list:
                    # print(v,"key is---------",k)
    #                 if any('.' in n for n in v):
                    super_dic[k] = list(map(float, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])
                else:
                    super_dic[k] = list(map(int, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])
            elif k in avg_insight:
                if type(v) == list:
                    # print(v,"key is---------",k)
                    # if any('.' in n for n in v):
                    super_dic[k] = list(map(float, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])/len(super_dic[k])
                else:
                    super_dic[k] = list(map(int, super_dic[k]))
                    super_dic[k] = sum(super_dic[k])/len(super_dic[k])
        print("impp are--------------------------------------------------",super_dic['impressions'])
        if super_dic['impressions']!=0:
            super_dic['ctr']=(super_dic['clicks']/super_dic['impressions'])*100
        else:
            super_dic['ctr']=0
        super_dic['spend']=ad_spend
        super_dic['conversions']=ad_conversions
        if super_dic['conversions']!=0:
            super_dic['cost_per_conversions']=(super_dic['spend']/super_dic['conversions'])
        else:
            super_dic['cost_per_conversions']=0

        print("clicks-------------------------------------",super_dic['clicks'])
        print("impresi-----------------------------------",super_dic['impressions'])
        print("ctr---------------------------------------------",super_dic['ctr'])
        print('the spend is----------------------------',ad_spend)
        try:
            print("the number is------------------------------",df[df['ad_id']==ad_id].index[0])
            df1['insights'][df1[df1['ad_id']==ad_id].index[0]] = json.dumps((super_dic))
        except:
            print("Not found ad_id ", ad_id, 'in ads_preprocessed')
    return df1
```
